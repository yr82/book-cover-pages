```python
# Launch the training script for textual inversion - book illustrations

! accelerate launch textual_inversion.py \
    --gradient_accumulation_steps=4 \
    --pretrained_model_name_or_path="runwayml/stable-diffusion-v1-5" \
    --train_data_dir="/content/book-illustrations" \
    --output_dir="/content/drive/MyDrive/sd-book-illustrations" \
    --learning_rate=1e-04 \
    --placeholder_token="<book-illustration>" \
    --initializer_token="illustration" \
    --learnable_property="style" \
    --center_crop \
    --train_batch_size=1 \
    --num_train_epochs=10 \
    --num_validation_images=4 \
    --validation_epochs=1 \
    --validation_prompt="A library at midnight under a starry sky, <book-illustration> style" \
    --report_to="wandb"
```
